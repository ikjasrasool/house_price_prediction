```python
columns_to_remove=['date','yr_renovated', 'street', 'city',
                   'statezip', 'country']
#df.drop(columns=['date'], inplace=True)
df.drop(columns=columns_to_remove, inplace=True)
```

```python
import scipy.stats as stats

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Calculate the Z-score for numeric columns
z_score = stats.zscore(df_numeric)

# Set the threshold for outliers
threshold = 3

# Print the size of the DataFrame before removing outliers
print("Size before removing outliers:", df.shape)

# Identify and store outliers
df_outliers = df[(z_score > threshold).any(axis=1)]

# Remove outliers based on the Z-score
df = df[(z_score <= threshold).all(axis=1)]

# Print the size of the DataFrame after removing outliers
print("Size after removing outliers:", df.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df)
df_scalar=pd.DataFrame(scalar.transform(df),columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x=df.drop('price', axis=1)
y=df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
```

```python
model=[
    ('Random Forest',RandomForestRegressor()),
    ('KNN',KNeighborsRegressor()),
    ('Linear Regression',LinearRegression()),
    ('Desion Tree',DecisionTreeRegressor())
]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
for name,model in model:
    print(f"Model Name: {name}")
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"MSE: {mean_squared_error(y_test, y_pred)}")
    print(f"R2: {r2_score(y_test, y_pred)}")
    print(f"MAE: {mean_absolute_error(y_test, y_pred)}")
```
